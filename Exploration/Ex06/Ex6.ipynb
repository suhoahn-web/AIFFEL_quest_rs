```python
import pandas as pd
import os

# 데이터를 읽어봅시다.
train_data = pd.read_table(os.getenv("HOME") + '/work/sentiment_classification/data/ratings_train.txt')
test_data = pd.read_table(os.getenv("HOME") + '/work/sentiment_classification/data/ratings_test.txt')

train_data.head()
```

```python
from konlpy.tag import Mecab
import numpy as np
from collections import Counter

tokenizer = Mecab()
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

def load_data(train_data, test_data, num_words=10000):
    train_data.drop_duplicates(subset=['document'], inplace=True)
    train_data = train_data.dropna(how = 'any')
    test_data.drop_duplicates(subset=['document'], inplace=True)
    test_data = test_data.dropna(how = 'any')

    X_train = []
    for sentence in train_data['document']:
        temp_X = tokenizer.morphs(sentence) # 토큰화
        temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
        X_train.append(temp_X)

    X_test = []
    for sentence in test_data['document']:
        temp_X = tokenizer.morphs(sentence) # 토큰화
        temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
        X_test.append(temp_X)

    words = np.concatenate(X_train).tolist()
    counter = Counter(words)
    counter = counter.most_common(10000-4)
    vocab = ['', '', '', ''] + [key for key, _ in counter]
    word_to_index = {word:index for index, word in enumerate(vocab)}

    def wordlist_to_indexlist(wordlist):
        return [word_to_index[word] if word in word_to_index else word_to_index[''] for word in wordlist]

    X_train = list(map(wordlist_to_indexlist, X_train))
    X_test = list(map(wordlist_to_indexlist, X_test))

    return X_train, np.array(list(train_data['label'])), X_test, np.array(list(test_data['label'])), word_to_index

X_train, y_train, X_test, y_test, word_to_index = load_data(train_data, test_data)
```

```python
# 문장 1개를 활용할 딕셔너리와 함께 주면, 단어 인덱스 리스트 벡터로 변환해 주는 함수입니다.
# 단, 모든 문장은 <BOS>로 시작하는 것으로 합니다.
def get_encoded_sentence(sentence, word_to_index):
    return [word_to_index['<BOS>']]+[word_to_index[word] if word in word_to_index else word_to_index['<UNK>'] for word in sentence.split()]

# 여러 개의 문장 리스트를 한꺼번에 단어 인덱스 리스트 벡터로 encode해 주는 함수입니다.
def get_encoded_sentences(sentences, word_to_index):
    return [get_encoded_sentence(sentence, word_to_index) for sentence in sentences]

# 숫자 벡터로 encode된 문장을 원래대로 decode하는 함수입니다.
def get_decoded_sentence(encoded_sentence, index_to_word):
    return ' '.join(index_to_word[index] if index in index_to_word else '<UNK>' for index in encoded_sentence[1:])  #[1:]를 통해 <BOS>를 제외

# 여러 개의 숫자 벡터로 encode된 문장을 한꺼번에 원래대로 decode하는 함수입니다.
def get_decoded_sentences(encoded_sentences, index_to_word):
    return [get_decoded_sentence(encoded_sentence, index_to_word) for encoded_sentence in encoded_sentences]
```

```python
# [Step 1] 데이터 길이 분포 확인 및 패딩 (Padding)
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np

# 1. 단어 사전에 특수 토큰 추가 (에러 방지용 필수 코드)
# 이미 있다면 덮어쓰고, 없다면 새로 추가합니다.
word_to_index["<PAD>"] = 0
word_to_index["<BOS>"] = 1
word_to_index["<UNK>"] = 2
word_to_index["<UNUSED>"] = 3

# 인덱스 -> 단어 사전도 동기화
index_to_word = {index:word for word, index in word_to_index.items()}

# 2. 문장 길이 분포 확인
total_data_text = list(X_train) + list(X_test)
num_tokens = [len(tokens) for tokens in total_data_text]
num_tokens = np.array(num_tokens)

print('문장길이 평균 : ', np.mean(num_tokens))
print('문장길이 최대 : ', np.max(num_tokens))
print('문장길이 표준편차 : ', np.std(num_tokens))

# 3. 최대 길이(maxlen) 설정
# (평균 + 2*표준편차)로 계산하면 약 41이 나옵니다.
max_tokens = np.mean(num_tokens) + 2 * np.std(num_tokens)
maxlen = int(max_tokens)
print('pad_sequences maxlen : ', maxlen)
print('전체 문장의 {}%가 maxlen 설정값 이내에 포함됩니다. '.format(np.sum(num_tokens < max_tokens) / len(num_tokens)))

# 4. 패딩(Padding) 적용
# 앞서 정의한 <PAD> 토큰(0번)을 사용하여 길이를 맞춥니다.
X_train = tf.keras.preprocessing.sequence.pad_sequences(X_train, value=word_to_index["<PAD>"], padding='pre', maxlen=maxlen)
X_test = tf.keras.preprocessing.sequence.pad_sequences(X_test, value=word_to_index["<PAD>"], padding='pre', maxlen=maxlen)

print("패딩 적용 완료!")
print("훈련 데이터 shape:", X_train.shape)
print("테스트 데이터 shape:", X_test.shape)
```

```python
# [Step 2] Gensim을 활용한 자체 Word2Vec 학습 및 분석 (루브릭 필수)
from gensim.models import Word2Vec
import os

# 1. 숫자 인덱스를 다시 단어로 바꾸는 역변환 함수 정의
def get_decoded_sentences(encoded_sentences, index_to_word):
    decoded_sentences = []
    for sentence in encoded_sentences:
        # 0:패딩, 1:BOS, 2:UNK, 3:UNUSED 는 제외하고 실제 단어만 복원합니다.
        decoded = [index_to_word[index] for index in sentence if index > 3]
        decoded_sentences.append(decoded)
    return decoded_sentences

# 2. 학습 데이터 복원 (숫자 -> 단어 리스트)
total_data_indices = list(X_train) + list(X_test)
total_data_text = get_decoded_sentences(total_data_indices, index_to_word)

# 3. Word2Vec 학습
print("Word2Vec 학습을 시작합니다...")

# [수정됨] 경로를 절대 경로에서 '현재 폴더'로 변경합니다.
word2vec_file_path = 'word2vec_ko.model' 

word2vec = Word2Vec(sentences=total_data_text, vector_size=100, window=5, min_count=5, workers=4, sg=0)

# 4. 학습된 임베딩 분석 (결과 확인)
print("------------------------------------------------")
print("'영화'와 유사한 단어:", word2vec.wv.most_similar("영화"))
print("'배우'와 유사한 단어:", word2vec.wv.most_similar("배우"))
print("------------------------------------------------")

# 5. 모델 저장
word2vec.save(word2vec_file_path)
print(f"Word2Vec 모델이 '{word2vec_file_path}'에 저장되었습니다!")
```

```python
# [Step 3] Word2Vec 가중치를 딥러닝 모델에 적용하기
vocab_size = 10000    # 어휘 사전의 크기(10,000개의 단어)
word_vector_dim = 100  # 워드 벡터의 차원 수 (위에서 만든 Word2Vec과 동일해야 함)

# 임베딩 행렬(Embedding Matrix) 준비
embedding_matrix = np.random.rand(vocab_size, word_vector_dim)

# embedding_matrix에 Word2Vec 워드 벡터를 단어 하나씩 차례대로 카피한다.
for word, i in word_to_index.items():
    if i < vocab_size: # vocab_size 범위 내의 인덱스만 사용
        if word in word2vec.wv:
            embedding_matrix[i] = word2vec.wv[word]
```

```python
# [Step 4] 3가지 모델 구성 및 학습 (루브릭 필수)
# 검증셋 분리 (전체 데이터의 약 20%~30%)
# validation set 30000건 분리
x_val = X_train[:30000]
y_val = y_train[:30000]

# validation set을 제외한 나머지
partial_x_train = X_train[30000:]
partial_y_train = y_train[30000:]

print(partial_x_train.shape)
print(partial_y_train.shape)
```

```python
# 모델 1: LSTM (가장 대중적인 모델)
from tensorflow.keras.initializers import Constant

model_lstm = tf.keras.Sequential()
model_lstm.add(tf.keras.layers.Embedding(vocab_size, 
                                 word_vector_dim, 
                                 embeddings_initializer=Constant(embedding_matrix),  # Word2Vec 적용
                                 input_length=maxlen, 
                                 trainable=True))   # Fine-tuning 허용
model_lstm.add(tf.keras.layers.LSTM(128))   # LSTM 유닛 수
model_lstm.add(tf.keras.layers.Dense(8, activation='relu'))
model_lstm.add(tf.keras.layers.Dense(1, activation='sigmoid'))  # 최종 출력

model_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_lstm.summary()

# 학습
history_lstm = model_lstm.fit(partial_x_train, partial_y_train, epochs=5, batch_size=512, validation_data=(x_val, y_val), verbose=1)

# 평가
results = model_lstm.evaluate(X_test,  y_test, verbose=2)
print("LSTM 정확도:", results[1])
```

```python
#모델 2: 1-D CNN (속도가 빠르고 성능이 좋음)
model_cnn = tf.keras.Sequential()
model_cnn.add(tf.keras.layers.Embedding(vocab_size, word_vector_dim, embeddings_initializer=Constant(embedding_matrix), input_length=maxlen, trainable=True))
model_cnn.add(tf.keras.layers.Conv1D(16, 7, activation='relu'))
model_cnn.add(tf.keras.layers.MaxPooling1D(5))
model_cnn.add(tf.keras.layers.Conv1D(16, 7, activation='relu'))
model_cnn.add(tf.keras.layers.GlobalMaxPooling1D())
model_cnn.add(tf.keras.layers.Dense(8, activation='relu'))
model_cnn.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model_cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_cnn.summary()

# 학습
history_cnn = model_cnn.fit(partial_x_train, partial_y_train, epochs=5, batch_size=512, validation_data=(x_val, y_val), verbose=1)

# 평가
results = model_cnn.evaluate(X_test,  y_test, verbose=2)
print("1-D CNN 정확도:", results[1])
```

```python
# 모델 3: GlobalMaxPooling1D (가장 단순한 구조)
model_gmp = tf.keras.Sequential()
model_gmp.add(tf.keras.layers.Embedding(vocab_size, word_vector_dim, embeddings_initializer=Constant(embedding_matrix), input_length=maxlen, trainable=True))
model_gmp.add(tf.keras.layers.GlobalMaxPooling1D())
model_gmp.add(tf.keras.layers.Dense(8, activation='relu'))
model_gmp.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model_gmp.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_gmp.summary()

# 학습
history_gmp = model_gmp.fit(partial_x_train, partial_y_train, epochs=5, batch_size=512, validation_data=(x_val, y_val), verbose=1)

# 평가
results = model_gmp.evaluate(X_test,  y_test, verbose=2)
print("GlobalMaxPooling 정확도:", results[1])
```

```python
# [Step 6] 사전학습 임베딩(Pre-trained) 로드 및 비교 분석 (루브릭 달성용)
import gensim
import os
import numpy as np
import pickle

# 1. 파일 확인
if not os.path.exists('ko.bin'):
    print("ko.bin 파일이 없습니다. 다운로드를 확인하세요.")
else:
    print("ko.bin 파일이 준비되었습니다.")

print("사전학습 모델 강제 복구를 시작합니다...")

pretrained_w2v = None

try:
    # [핵심] Gensim load() 대신 pickle로 강제 로드
    with open('ko.bin', 'rb') as f:
        # 인코딩 문제 방지를 위해 latin1 옵션 사용
        model_data = pickle.load(f, encoding='latin1')
        
    print("▶ Pickle 로드 성공! 데이터 구조를 변환합니다...")
    
    # [복구] 구버전 데이터 구조(syn0, vocab)를 신버전 KeyedVectors로 이식
    # 구버전 모델은 model_data가 곧 객체이거나 딕셔너리 형태일 수 있음
    
    # 1. 빈 KeyedVectors 객체 생성 (벡터 크기: 200)
    # Kyubyong 모델은 200차원입니다.
    vector_size = 200
    pretrained_w2v = gensim.models.KeyedVectors(vector_size)
    
    # 2. 데이터 추출 (구버전 속성 이름: syn0, index2word 등)
    if hasattr(model_data, 'syn0'):
        vectors = model_data.syn0
    elif hasattr(model_data, 'wv') and hasattr(model_data.wv, 'vectors'):
        vectors = model_data.wv.vectors
    else:
        # 속성을 직접 찾기 어려우면 dir()로 확인해볼 수도 있음
        vectors = model_data.syn0 # 보통 구버전은 이거임
        
    if hasattr(model_data, 'index2word'):
        index2word = model_data.index2word
    elif hasattr(model_data, 'wv') and hasattr(model_data.wv, 'index_to_key'):
        index2word = model_data.wv.index_to_key
    else:
        index2word = model_data.index2word
        
    # 3. 신버전 객체에 주입
    pretrained_w2v.add_vectors(index2word, vectors)
    print(f"▶ 복구 완료! 단어 개수: {len(index2word)}, 차원: {vector_size}")

except Exception as e:
    print(f"▷ 복구 실패: {e}")
    # 만약 위 방법도 안 되면, 정말 호환 불가하므로 최후의 수단: FastText 모델 다운로드 제안
    print("오래된 모델 복구에 실패했습니다. 최신 FastText 모델 사용을 권장합니다.")
    raise


# 3. 비교 분석
print("\n[비교 결과]")
# 자체 학습 모델(word2vec) 확인
try:
    print(f"[자체학습] '영화' 유의어: {word2vec.wv.most_similar('영화', topn=5)}")
except:
    pass

def get_similar_words(kv, word):
    try:
        return kv.most_similar(word, topn=5)
    except KeyError:
        return ["사전에 없는 단어"]

print(f"[사전학습] '영화' 유의어: {get_similar_words(pretrained_w2v, '영화')}")
print(f"[사전학습] '배우' 유의어: {get_similar_words(pretrained_w2v, '배우')}")
print(f"[사전학습] '사랑' 유의어: {get_similar_words(pretrained_w2v, '사랑')}")


# 4. 딥러닝 모델에 이식
print("\n임베딩 매트릭스 생성 중...")
vocab_size = 10000
word_vector_dim = 200

embedding_matrix = np.random.rand(vocab_size, word_vector_dim)

count = 0
for word, i in word_to_index.items():
    if i < vocab_size:
        if word in pretrained_w2v:
            embedding_matrix[i] = pretrained_w2v[word]
            count += 1

print(f"\n[최종 결과] 전체 {vocab_size}개 단어 중 {count}개의 단어를 사전학습 모델에서 가져왔습니다.")
```
