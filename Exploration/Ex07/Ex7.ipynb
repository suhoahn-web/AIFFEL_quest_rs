```python
import torch
from torch.nn.utils.rnn import pad_sequence

class Tokenizer:
    def __init__(self, filters=''):
        self.word_index = {}
        self.index_word = {}
        self.filters = filters

    def fit_on_texts(self, corpus):
        # 각 문장을 토큰화하여 단어 사전을 생성 (인덱스는 1부터 시작)
        for sentence in corpus:
            tokens = sentence.split() if isinstance(sentence, str) else sentence
            for token in tokens:
                if token not in self.word_index:
                    self.word_index[token] = len(self.word_index) + 1
        # 역방향 사전 생성
        self.index_word = {idx: word for word, idx in self.word_index.items()}

    def texts_to_sequences(self, corpus):
        sequences = []
        for sentence in corpus:
            tokens = sentence.split() if isinstance(sentence, str) else sentence
            # 존재하지 않는 단어는 0으로 처리
            seq = [self.word_index.get(token, 0) for token in tokens]
            sequences.append(torch.tensor(seq, dtype=torch.long))
        return sequences

    def sequences_to_texts(self, sequences):
        texts = []
        for seq in sequences:
            # tensor인 경우 리스트로 변환
            if isinstance(seq, torch.Tensor):
                seq = seq.tolist()
            # 패딩 토큰(0)은 제외하고 디코딩
            tokens = [self.index_word.get(idx, "") for idx in seq if idx != 0]
            texts.append(tokens)
        return texts
```

```python
def tokenize(corpus):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(corpus)
    sequences = tokenizer.texts_to_sequences(corpus)
    # padding 토큰은 0
    tensor = pad_sequence(sequences, batch_first=True, padding_value=0)
    return tensor, tokenizer
```

```python
# Step 3 Tokenizer 함수 작성
def sp_tokenize(s, corpus):

    tensor = []

    for sen in corpus:
        tensor.append(s.EncodeAsIds(sen))

    with open("./korean_spm.vocab", 'r') as f:
        vocab = f.readlines()

    word_index = {}
    index_word = {}

    for idx, line in enumerate(vocab):
        word = line.split("\t")[0]

        word_index.update({word:idx})
        index_word.update({idx:word})

    tensor = pad_sequence(tensor, batch_first=True, padding_value=0)

    return tensor, word_index, index_word
```

```python
# Step 4 네이버 영화리뷰 감정 분석에 SentencePiece 적용
import pandas as pd
import urllib.request
import sentencepiece as spm
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torch.nn.utils.rnn import pad_sequence

# ---------------------------------------------------------
# 1. 데이터 준비 & SentencePiece 학습
# ---------------------------------------------------------
# GPU 사용 가능 여부 확인
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"현재 사용 장치: {device}")

# 데이터 다운로드 및 전처리
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt", filename="ratings_train.txt")
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings_test.txt", filename="ratings_test.txt")

train_df = pd.read_table('ratings_train.txt').dropna()
test_df = pd.read_table('ratings_test.txt').dropna()

# SentencePiece 학습용 텍스트 파일 생성
with open('nsmc.txt', 'w', encoding='utf-8') as f:
    f.write('\n'.join(train_df['document'].astype(str)))

# SentencePiece 학습 (vocab_size=8000)
vocab_size = 8000
spm.SentencePieceTrainer.Train(
    f'--input=nsmc.txt --model_prefix=nsmc_spm --vocab_size={vocab_size} '
    f'--model_type=unigram --pad_id=0 --bos_id=1 --eos_id=2 --unk_id=3'
)

# Tokenizer 로드
sp = spm.SentencePieceProcessor()
sp.Load('nsmc_spm.model')

# ---------------------------------------------------------
# 2. PyTorch용 데이터셋 정의
# ---------------------------------------------------------
class NSMCDataset(Dataset):
    def __init__(self, df, sp, max_len=50):
        self.sentences = df['document'].tolist()
        self.labels = df['label'].tolist()
        self.sp = sp
        self.max_len = max_len

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        # 토큰화 및 정수 인코딩
        encoded = self.sp.EncodeAsIds(self.sentences[idx])
        # 길이 제한 (max_len 보다 길면 자름)
        if len(encoded) > self.max_len:
            encoded = encoded[:self.max_len]
        
        return torch.tensor(encoded), torch.tensor(self.labels[idx])

# Collate Fn: 배치마다 길이를 맞춰주는 패딩 함수
def collate_fn(batch):
    inputs, labels = zip(*batch)
    # 패딩 (padding_value=0)
    padded_inputs = pad_sequence(inputs, batch_first=True, padding_value=0)
    return padded_inputs, torch.stack(labels)

# 데이터셋 & 데이터로더 생성
train_dataset = NSMCDataset(train_df, sp)
test_dataset = NSMCDataset(test_df, sp)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True, collate_fn=collate_fn)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False, collate_fn=collate_fn)

# ---------------------------------------------------------
# 3. LSTM 모델 설계 (PyTorch)
# ---------------------------------------------------------
class LSTMClassifier(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim):
        super(LSTMClassifier, self).__init__()
        # padding_idx=0 설정 중요 (패딩은 학습하지 않음)
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        embedded = self.embedding(x)
        # LSTM 출력 중 마지막 타임스텝의 은닉 상태만 사용
        _, (hidden, _) = self.lstm(embedded)
        # hidden[-1]은 마지막 레이어의 마지막 시점 상태
        out = self.fc(hidden[-1])
        return self.sigmoid(out)

model = LSTMClassifier(vocab_size=vocab_size, embedding_dim=100, hidden_dim=128).to(device)
criterion = nn.BCELoss() # 이진 분류 손실함수
optimizer = optim.Adam(model.parameters(), lr=0.001)

# ---------------------------------------------------------
# 4. 학습 및 평가 루프
# ---------------------------------------------------------
print("학습 시작...")
for epoch in range(5): # 5 Epoch만 진행
    model.train()
    total_loss = 0
    
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.float().to(device)
        
        optimizer.zero_grad()
        outputs = model(inputs).squeeze()
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        
    print(f"Epoch {epoch+1} 완료 | 평균 Loss: {total_loss / len(train_loader):.4f}")

# 테스트 평가
model.eval()
correct = 0
total = 0

with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.float().to(device)
        outputs = model(inputs).squeeze()
        predicted = (outputs > 0.5).float()
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"\n최종 테스트 정확도: {100 * correct / total:.2f}%")

# 실제 문장 테스트 함수
def predict_sentiment(sentence):
    model.eval()
    encoded = sp.EncodeAsIds(sentence)
    tensor = torch.tensor(encoded).unsqueeze(0).to(device) # 배치 차원 추가
    score = model(tensor).item()
    
    if score > 0.5:
        print(f"'{sentence}' -> 긍정 ({score*100:.2f}%)")
    else:
        print(f"'{sentence}' -> 부정 ({(1-score)*100:.2f}%)")

predict_sentiment("이 영화 진짜 재밌다")
predict_sentiment("시간 아까워 죽는 줄")
```

```python
#SentencePiece 성능 다각도 비교(Unigram vs BPE)
import sentencepiece as spm
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torch.nn.utils.rnn import pad_sequence

# --- 실험을 위한 함수화 (설정을 바꿔가며 돌리기 위해) ---
def run_experiment(model_type, vocab_size):
    print(f"\n======== [실험 시작] Type: {model_type} / Vocab: {vocab_size} ========")
    
    # 1. SP 모델 학습
    prefix = f'nsmc_{model_type}_{vocab_size}'
    spm.SentencePieceTrainer.Train(
        f'--input=nsmc.txt --model_prefix={prefix} --vocab_size={vocab_size} '
        f'--model_type={model_type} --pad_id=0 --bos_id=1 --eos_id=2 --unk_id=3'
    )
    
    # 2. 토크나이저 로드
    sp_exp = spm.SentencePieceProcessor()
    sp_exp.Load(f'{prefix}.model')
    
    # 3. 데이터셋 재생성 (새로운 토크나이저 적용)
    # (NSMCDataset 클래스는 이전 코드에 정의된 것을 그대로 씁니다)
    train_ds = NSMCDataset(train_df, sp_exp)
    test_ds = NSMCDataset(test_df, sp_exp)
    
    train_ld = DataLoader(train_ds, batch_size=64, shuffle=True, collate_fn=collate_fn)
    test_ld = DataLoader(test_ds, batch_size=64, shuffle=False, collate_fn=collate_fn)
    
    # 4. 모델 초기화 (동일한 조건)
    model = LSTMClassifier(vocab_size=vocab_size, embedding_dim=100, hidden_dim=128).to(device)
    criterion = nn.BCELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    
    # 5. 간단하게 3 Epoch만 학습 (비교 목적이므로)
    model.train()
    for epoch in range(3):
        for inputs, labels in train_ld:
            inputs, labels = inputs.to(device), labels.float().to(device)
            optimizer.zero_grad()
            outputs = model(inputs).squeeze()
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
    # 6. 평가
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in test_ld:
            inputs, labels = inputs.to(device), labels.float().to(device)
            outputs = model(inputs).squeeze()
            predicted = (outputs > 0.5).float()
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            
    acc = 100 * correct / total
    print(f"======== [실험 종료] 정확도: {acc:.2f}% ========")
    return acc

# --- 비교 실험 실행 ---
# 이미 Unigram(8000)은 해봤으니, BPE(8000)과 비교해 봅니다.
results = {}

# 실험 1: Unigram (이미 했지만 공정한 비교를 위해 짧게 다시 실행)
results['Unigram'] = run_experiment(model_type='unigram', vocab_size=8000)

# 실험 2: BPE (Byte Pair Encoding)
results['BPE'] = run_experiment(model_type='bpe', vocab_size=8000)

# --- 최종 결과 출력 ---
print("\n[최종 비교 결과]")
for key, value in results.items():
    print(f"{key} 모델 정확도: {value:.2f}%")
```
